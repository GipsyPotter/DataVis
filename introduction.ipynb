**Bài 1**
Tự tạo một Dataframe ngẫu nhiên có những cột sau đây

Name: tự chọn 

Age: 18 - 60 tuổi

Math: 0.0 - 10.0

English: 0.0 - 10.0

History: 0.0 - 10.0

Note: Các bạn có thể tự tạo bằng hàm random (cộng điểm) hoặc tự thêm thủ công.

a) Tạo Dataframe có ít nhất 10 hàng.

b) Thêm 1 hàng dữ liệu (thông tin bản thân)

c) Thêm cột có tên là Result.

Result: Nếu (Math + English + History)/3 >= 5 thì PASS ngược lại thì NOT PASS

Note: Có thể dừng lại ở bước tính điểm TB

d) xuất ra file csv có tên là result.csv

```python
import pandas as pd
import numpy as np
# a)
num_records = 101

names = ["Peter", "Potter", "Potterio", "Steidernman", "Thu", "DoubleDev", "Tom", "Katio", "Eric", "David"]
name_column = np.random.choice(names, size=num_records)

age_column = np.random.randint(18, 61, size=num_records)

math_column = np.random.uniform(0.0, 10.0, size=num_records)
english_column = np.random.uniform(0.0, 10.0, size=num_records)
history_column = np.random.uniform(0.0, 10.0, size=num_records)

df = pd.DataFrame({"Name": name_column, "Age": age_column, "Math": math_column, "English": english_column, "History": history_column})

df[["Math", "English", "History"]] = df[["Math", "English", "History"]].round(1)

print(df)
# b)
my_row = pd.DataFrame({
    "Name": ["Porter"],
    "Age": [16],
    "Math": [9.0],
    "English": [10],
    "History": [10]
})

df = df.append(my_row, ignore_index=True)
print(df)

# c)
df["Result"] = ["PASS" if (row["Math"] + row["English"] + row["History"]) / 3 >= 5 else "NOT PASS" for index, row in df.iterrows()]

print(df)

# d)

df.to_csv("result.csv")
```

**Bài 2**

a) Xử lý file csv sau bằng những kiến thức đã học (hạn chế xoá)

b) Thống kê dữ liệu sau (tất cả các cột)

Ví dụ:

- Bao nhiêu huy chương đồng, bạc, vàng?

- Mode, Median, Standard Deviation, Mean bằng bao nhiêu?...

* Lưu ý: Chỉ cần in ra màn hình thông tin tương ứng

```python
import numpy as np
import pandas as pd
import random

df = pd.read_csv("http://winterolympicsmedals.com/medals.csv")
# [KHÔNG SỬA CODE]
# percentage of NA to add
n = 20 

# generate random numbers between 0 and 1 for each row
rand_nums = [random.random() for _ in range(len(df))]

# replace values in column with NA where random number is less than n/100
df.loc[np.array(rand_nums) < n/100, "City"] = np.nan
df.loc[np.array(rand_nums) < n/100, "Year"] = "UNKNOWN"
print(df)
# [KHÔNG SỬA CODE]


# [CODE Ở ĐÂY TRỞ XUỐNG]

#CLEANING
df = df.fillna(method="ffill").fillna(method="bfill")
df["Year"] = df["Year"].replace("UNKNOWN", np.nan).ffill()
df["Year"] = df["Year"].replace("UNKNOWN", np.nan).bfill()
df["Year"] = df["Year"].astype(int)
print(df)

#Count all medal
medal_counts = df.groupby(["NOC", "Medal"])["Medal"].count()
print(medal_counts)

#Cal Year
mean_year = df["Year"].mean().round(0)
median_year = df["Year"].median().round(0)
mode_year = df["Year"].mode().round(0)
std_year = df["Year"].std().round(0)

print("Mean:", mean_year)
print("Median:", median_year)
print("Mode:", mode_year)
print("STD:", std_year)
```
