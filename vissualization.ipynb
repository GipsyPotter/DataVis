```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# generate random data for light intensity
np.random.seed(42)
light_data = np.random.normal(10, 3, 1000)

# create a histogram using seaborn
sns.histplot(light_data, kde=False, color='blue')

# set the x and y labels
plt.xlabel('Light Intensity')
plt.ylabel('Frequency')

# display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Create the heatmap
sns.heatmap(data=flights.pivot('month', 'year', 'passengers'))

# Show the plot
plt.show()
```
